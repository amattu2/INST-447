# Data Analysis

## Visualization 1

```python
# Subset by department
temp = df.loc[df['department'] == 'deli']

# Build Plot
sns.barplot(x = 'order_hour_of_day', y = 'num_orders_hour', hue = 'tot_orders_dept', data = temp)
plt.title("Barplot of Deli Orders by Hour")
plt.ylabel("Order Count")
plt.xlabel("Hour (24)")

# Hide Output
None;
```

### Analysis
From this above graph we can see a clear upward trend in Deli-type orders beginning at 6 AM. This trend continues until approximately 4:00 PM, when the number of orders begins to decrease somewhat rapidly. The decrease continues until the next day at about 4 AM, when the repeating uptick in sales repeats itself. This graph provides clear evidence that traditional deli sales occur during normal hours of the day.

## Visualization 2

```python
# Subset by department
temp = df.loc[df['department'] == 'alcohol']

# Build Plot
sns.barplot(x = 'order_hour_of_day', y = 'num_orders_hour', hue = 'tot_orders_dept', data = temp)
plt.title("Barplot of Alcoholic Orders by Hour")
plt.ylabel("Order Count")
plt.xlabel("Hour (24)")

# Hide Output
None;
```

### Analysis
The above barplot outlines the sales trend for alcoholic-type beverage orders collected from Instacart's public dataset. Similar to the Deli trend, we can see a significant uptick in sales beginning around 9 AM, with the peak sales hitting at 3:00 PM every day. Surprisingly, the sales drastically drop down to nearly 0 throughout the next ~12 hours, with the minimal amount of sales being 3:00 AM to 4:00 AM every morning. This graph suggests that alcoholic beverage sales are much more common towards the end of the day, as opposed to deli sales (seen previously).

## Visualization 3

```python
# Subset by department
temp = df.loc[df['department'] == 'snacks']

# Build Plot
sns.barplot(x = 'order_hour_of_day', y = 'num_orders_hour', hue = 'tot_orders_dept', data = temp)
plt.title("Barplot of Snack Orders by Hour")
plt.ylabel("Order Count")
plt.xlabel("Hour (24)")

# Hide Output
None;
```

### Analysis
The above graph details the Instacart orders categorized by the snack department in stores. While it does have a similar trend to the deli sales graph, the snacks department appears to maintain customer "attention" for a longer duration throughout the day. Moreover, the orders don't drop significantly until around 2:00 AM every morning, but leading up to this drop is a simply a gradual decline in sales. This suggests that people are ordering snacks generally throughout the day, even through early morning.
