# Beautiful Soup

```python
# example html (this is not yet in beautiful soup html format. This is just a string that includes html syntax)
html = """<!DOCTYPE html>
            <html>
            <head>
            <title>Page Title</title>
            </head>
            <body>

            <h1>This is a Heading</h1>
            <p>This is first paragraph.</p>
            <p>This is second paragraph.</p>
            <p>This is third paragraph.</p>

            </body>
            </html>
        """
```

```python
# convert the example html string into beautiful soup html
htmlObj = BeautifulSoup(html, "html.parser")

# print the beautifulsoup html object (it should print the html)
print(htmlObj)
```

```python
# use find to get the title of the example html website
title = htmlObj.find("title")

# print the text within the title tags (i.e. the tags should not be included in the print) 
title.text
```

```python
# use findall to get the level 1 headings (h1) 
headings = htmlObj.findAll("h1")

# print the text within the heading tags (i.e. the tags should not be included in the print) (HINT: since you are using findall, you must print using a for loop)
for heading in headings:
  print(heading.text)
```

```python
# use findall to get the paragraphs 
paragraphs = htmlObj.findAll("p")

# print the text within the paragraph tags (i.e. the tags should not be included in the print) (HINT: since you are using findall, you must print using a for loop)
for pg in paragraphs:
  print(pg.text)
```

## Using beautiful soup to get weather data

```python
# Use the requests module to download the content/html (text) from https://darksky.net/zipcode/20742/us12/en
page = requests.get("https://darksky.net/zipcode/20742/us12/en")

page.status_code
```

```python
# convert the html from the darksky website into a beautiful soup object
soup = BeautifulSoup(page.text, 'html.parser')

# print the beautifulsoup object created from the darksky website
print(soup)
```

```python
# get and save the beautifulsoup elements that have the predicted minimum temperatures for each day
# HINT: use element inspector on the darksky website to notice where in the HTML the minimum temperature is stored
# HINT: look for a specific class of span elements
elements = soup.findAll('span', {'class':'minTemp'})

# print the elements that contain the minimum temperatures for each day (use a for loop to print the minimum temps)
for i in elements:
  print(i.getText())
```

```python
# get and save the beautifulsoup elements that have the predicted maximum temperatures for each day
# HINT: use element inspector on the darksky website to notice where in the HTML the maximum temperature is stored
# HINT: look for a specific class of span elements
elements = soup.findAll('span', {'class':'maxTemp'})

# print the elements that contain the maximum temperatures for each day (use a for loop to print the maximum temps)
for i in elements:
  print(i.getText())
```
