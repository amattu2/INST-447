```python
import pandas as pd
import numpy as np
import requests
from pandas.io.json import json_normalize #special package in pandas
import json
```

# Part A: APIs

## Wiki Dog Info

```python
# download the response from Wikipedia API endpoint https://en.wikipedia.org/w/api.php 
# to get general information about the Dog Wikipedia article
# ask for the output as json
# HINT: prop = info
data = requests.get("https://en.wikipedia.org/w/api.php?action=query&prop=info&titles=Dog&format=json")

# save output and print
print(data.text)
```

```python
# convert json string into json object using json.loads (HINT: you need the text from the request object)
d = json.loads(data.text)
d
```

## Wiki Dog Revisions

```python
# download the response from Wikipedia API endpoint https://en.wikipedia.org/w/api.php 
# to get the latest revision information (last revision) for the Dog Wikipedia article
# ask for the output as json
# HINT: prop = revisions
data = requests.get("https://en.wikipedia.org/w/api.php?action=query&prop=revisions&titles=Dog&format=json")

# save output and print
print(data.text)
```

```python
# convert json string into json object using json.loads (HINT: you need the text from the request object)
d = json.loads(data.text)
d
```

```python
# flatten json data into a data frame where each row is a revision (HINT: you should end up with a dataframe with one
# observation)
df = pd.json_normalize(d['query']['pages']['4269567']['revisions'])

# print dataframe
df.head()
```

```python
# download the last 15 revisions to the dog page
# ask for the output as json
# HINT: use rvlimit
data = requests.get("https://en.wikipedia.org/w/api.php?action=query&prop=revisions&titles=Dog&format=json&rvlimit=15")

# save output and print
data.text
```

```python
# convert json string into json object using json.loads (HINT: you need the text from the request object)
d = json.loads(data.text)
d
```

```python
# flatten json data into a data frame where each row is a revision 
df = pd.json_normalize(d['query']['pages']['4269567']['revisions'])

# count the number of rows
print(df.shape[0])

# print the first few rows
df.head()
```
